# 传统机器学习模型+数据基础强化
## 专有名词介绍
* 机器学习

机器学习（Machine Learning, ML）是人工智能（AI）的一个分支，其核心是通过算法和统计模型，使计算机系统能够从数据中自动“学习”并改进性能，而无需显式编程。

机器学习通过分析大量数据，识别其中的模式或规律，并利用这些模式进行预测、分类、决策或生成新内容。其性能随着数据量的增加或经验的积累而提升。机器学习程序从经验E中学习任务T，并用性能指标P衡量，若在T上的P随E提升而改善，则称其具有学习能力。
* 机器学习分类

* 点击率

点击率(Click-Through Rate, CTR)是数字营销和在线广告中的一个重要指标，用于衡量用户看到某个内容后实际点击的比例。

* 特征和标签

**特征**：指用于描述样本的属性或变量，是模型输入的原始数据（例如表格中的列）。

**标签**：指机器学习模型要预测的目标值（监督学习中的输出变量）。

* 常见的损失函数

损失函数（Loss Function）或目标函数（Objective Function）用于量化模型预测值与真实值之间的差异。它的目的是衡量模型表现的好坏。

* 逻辑回归

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

逻辑回归（Logistic Regression）是一种常用的分类算法，尽管其名称中包含“回归”二字，但它主要用于解决分类问题。它通过将线性回归的输出值映射到0到1之间，从而表示某个事件发生的概率。
* 决策树
* 决策树

https://scikit-learn.org/stable/modules/tree.html#tree

构建决策树的过程是一个递归的、自上而下的过程，目标是找到最佳的特征和分裂点，使得每个分支下的数据尽可能“纯净”。“纯净”指的是在分类任务中，叶节点中的样本尽可能属于同一个类别。一个样本从根节点开始，根据每个节点的问题和分支条件，逐步向下遍历，直到到达某个叶节点，从而得到它的分类或预测结果。

* 集成学习

https://scikit-learn.org/stable/modules/ensemble.html#ensemble

集成学习（Ensemble Learning）是一种强大的机器学习策略，它的核心思想是结合多个学习器（或称基模型、弱学习器）的预测结果，以获得比单一学习器更优越的性能。


Bagging（套袋法）通过自助采样法（Bootstrap Sampling）从原始数据集中有放回地抽取多个子集，每个子集用于训练一个基学习器。由于每次抽样都可能不同，每个基学习器看到的训练数据略有差异，从而产生不同的模型。

Boosting（提升法）训练一个元学习器（Meta-learner）或二级学习器（Level-1 learner）来组合所有基学习器（一级学习器，Level-0 learners）的预测结果。基学习器的输出作为元学习器的输入。

* 随机森林

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

随机森林（Random Forest）属于集成学习中的 Bagging（套袋法） 类型。它由多棵独立的决策树组成的“森林”，每棵树都对问题做出自己的判断，最终的结果由这些树的共同决定（例如，多数投票或平均）。

* XGBoost

https://xgboost.readthedocs.io/en/latest/tutorials/model.html

XGBoost 全称为 “Extreme Gradient Boosting”（极限梯度提升），它是一个高度优化、分布式且灵活的库，实现了梯度提升（Gradient Boosting）算法。梯度提升本身是一种强大的集成学习技术。
!
XGBoost 的目标是将机器学习的原理与系统优化的考虑深度结合，以提供一个可扩展（scalable）、可移植（portable）和准确（accurate）的机器学习库。它在数据科学竞赛和工业界应用中都取得了巨大的成功，成为最受欢迎的算法之一

* KMeans

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans

K-Means 是一种简单但非常流行的无监督学习算法，主要用于解决聚类问题。它的目标是将数据集中的 N 个数据点划分到 K 个簇（cluster）中，使得同一个簇内的数据点相似度高，而不同簇之间的数据点相似度低。这里的“相似度”通常通过数据点到簇中心的距离来衡量。


* PCA

https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

主成分分析（Principal Component Analysis，PCA）是一种非常常见的降维技术。它属于无监督学习，主要目标是将高维数据转换成低维数据，同时尽可能保留原始数据中的重要信息。
PCA 的核心思想是找到数据中方差最大的方向，这些方向被称为主成分（Principal Components）。方差越大，说明数据在这个方向上的差异性越大，包含的信息就越多。
